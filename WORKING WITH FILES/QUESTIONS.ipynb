```python
Month vs Killed
Send Feedback
Given file "year2017.csv"
It is an open-source database including information on terrorist attacks around the world in 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period
Problem Statement :
Find the total number people killed in each month.
Print the month and count of killed people as integer value.
Output Format :
Month_1 Count_1
Month_2 Count_2
Month_3 Count_3
and so on
```

```python
## Open and read data file as specified in the question
## Print the required output in given format
import csv
with open("year2017.csv") as myfile:
    data=csv.DictReader(myfile,skipinitialspace=True) #object of csv class.
    
    a=list(data) #makes a list of dictionary.
    
    
    dic_of_month={} #dic with month as key and people killed as value
    
    
    
        
    for row in a:
        month=row["Month"]#1
        killed=row["Killed"]#'0.0'
        
        if killed!="":
            killed=float(killed)
            killed=int(killed)
        
            if month in dic_of_month:
                dic_of_month[month]+=killed
            
            else:
                dic_of_month[month]=killed
    #print(dic_of_month)
    for key in dic_of_month:
        print(key, dic_of_month[key])
```

```python
Country vs Casualties
Send Feedback
Given file "year2017.csv"
It is an open-source database including information on terrorist attacks around the world in 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period
Problem Statement :
Find total number of casualties in each country.
Print number of casualties as integer value.
Note : For any missing value of casualty column, take 0 as value
Output Format :
country_1 casualties_1
country_2 casualties_2
country_3 casualties_3
and so on
```

```python
## Open and read data file as specified in the question
## Print the required output in given format
## Open and read data file as specified in the question
## Print the required output in given format
import csv
with open("year2017.csv") as myfile:
    data=csv.DictReader(myfile,skipinitialspace=True) #object of csv class.
    
    a=list(data) #makes a list of dictionary.
    
    
    dic={} #dic with month as key and people killed as value
    #tup=()
    li=[]
    
    
    
        
    for row in a:
        #print(row)
        month=row["Country"]#1
        killed=row["casualities"]#'0.0'
        
        if killed!="":
            killed=float(killed)
            killed=int(killed)
            
        
            if month in dic:
                dic[month]+=killed
            
            else:
                dic[month]=killed
                li.append([month])
                
        else:
            if month in dic:
                dic[month]+=0
            else:
                dic[month]=0
                
    #print(dic)
    for ele in li:
        con=ele[0]
        print(con,dic[con])
    
```
